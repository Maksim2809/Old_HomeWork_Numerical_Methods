```python
import math
import numpy as np
```

```python
def f(x):
    return (0.6*x+1.5)**(1/2)/(2*x+((x**2)+3)**(1/2))

def f2(x):
    return math.sin(0.8*(x**2)+0.3)/(0.7+math.cos(1.2*x+0.3))
```
